```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Sample data
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 100)
y2 = np.cos(x2)

# Create shared y-axis plots
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 5))

ax1.plot(x1, y1)
ax1.set_title('Sine Curve')

ax2.plot(x2, y2)
ax2.set_title('Cosine Curve')

# Shared y-axis label
fig.text(0.04, 0.5, 'Y-Axis Label', va='center', rotation='vertical')

plt.tight_layout()
plt.show()
```

```python
# Overlayed on top of each other
plt.style.use('ggplot')
# sns.set_style('ticks')
sns.lineplot(x1, y1)
ax2 = plt.gca().twinx()

sns.lineplot(x2, y2, ax=ax2)
```
