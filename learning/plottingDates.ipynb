```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
test_df = pd.DataFrame({
        'dates': pd.date_range(start='2015-05-01', end='2015-06-01', freq='H'),
        'values': np.linspace(0, 100, 745)
    })

test_df
```
